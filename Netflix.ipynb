````python
df.fillna({
    'director': 'Not Available',
    'cast': 'Not Available',
    'country': 'Unknown'
}, inplace=True)
df.fillna
````

````python
# Fix data column
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
````

````python
df['date_added'].dtype
````

````python
df['year_added'] = df['date_added'].dt.year
df[['date_added', 'year_added']].head()
````

````python
df['main_country'] = df['country'].apply(lambda x: x.split(',')[0])
````

````python
df['main_genre'] = df['listed_in'].str.split(',').str[0].str.strip()
````

````python
# Duration Split
df['duration_value'] = df['duration'].str.extract('(\d+)').astype(float)
````

````python
df.head()
df.info()
````

````python
# Movies Vs TV Count
df['type'].value_counts()
````

````python
# Movies Vs TV Shows
import matplotlib.pyplot as plt

df['type'].value_counts().plot(kind='bar')
plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
````

````python
# Content Growth by release Year
yearly_content = df['release_year'].value_counts().sort_index()

yearly_content.plot(kind='line')
plt.title("Content Release Over Years")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.show()
````

````python
# Top 10 Countries
top_countries = df['main_country'].value_counts().head(10)

top_countries.plot(kind='bar')
plt.title("Top 10 Countries on Netflix")
plt.xlabel("Country")
plt.ylabel("Total Titles")
plt.show()
````

````python
# Top Genres
top_genres = df['main_genre'].value_counts().head(10)
top_genres.plot(kind='bar')
plt.title("Top Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
````

````python
# Rating Distributions
rating_count = df['rating'].value_counts()

rating_count.plot(kind='bar')
plt.title("Rating Distribution on Netflix")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
````
